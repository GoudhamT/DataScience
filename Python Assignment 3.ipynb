## 1.1 Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce()

```python
# find the sum of below list
def myreduce():
    lst = [2,4,6,8,90,34,123]
    sum_output = 0
    for no in lst:
        sum_output += no
    print(sum_output)
myreduce()
```

```python
#  Reduce function will do the same
from functools import reduce
reduce(lambda x,y:x+y,lst)
```

## 1.2 Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter()

```python
# find the even numbers for ranging 0 to 50
def myfilter(n):
    if n%2 == 0:
        return True
```

```python
odd_nos = []
for no in range(50):
    if myfilter(no):
        odd_nos.append(no)
print(odd_nos)
```

```python
#  Same scenario can be achived using filter function
list(filter(lambda x:x%2 == 0, range(50)))
```

## 2. Implement List comprehensions to produce the following lists.
## Write List comprehensions to produce the following Lists

## ['x', 'xx', 'xxx', 'xxxx', 'y', 'yy', 'yyy', 'yyyy', 'z', 'zz', 'zzz', 'zzzz']


```python
letter = ['x','y','z']
lst = [oneletter* no for oneletter in letter for no in range(1,5)]
print(lst)
```

## ['x', 'y', 'z', 'xx', 'yy', 'zz', 'xxx', 'yyy', 'zzz', 'xxxx', 'yyyy', 'zzzz']

```python
lst = [oneletter*no for no in range(1,5) for oneletter in letter]
print(lst)
```

## [(1, 1), (2, 1), (3, 1), (1, 2), (2, 2), (3, 2), (1, 3), (2, 3), (3, 3)]

```python
my_no_lst = [(x,y) for y in range(1,4) for x in range(1,4)]
print(my_no_lst)
```
